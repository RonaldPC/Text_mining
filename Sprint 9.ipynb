#### Conversión del texto a minúsculas (podría ser a mayúsculas también) para que una misma palabra no sea diferenciada por ser mayúscula o minúscula.

```python
Reagan_speech = Reagan_speech.lower()
print(Reagan_speech)
```

#### Tokenozación para dividir el texto en unidades más pequeñas (palabras, signos de puntuación...)

```python
from nltk.tokenize import word_tokenize

Reagan_speech_tokens = word_tokenize(Reagan_speech)
print(Reagan_speech_tokens)
```

#### Eliminación de los signos de puntuación

```python
Reagan_speech_tkns_NoSigPunt = [tkn for tkn in Reagan_speech_tokens if tkn.isalpha()]
print(Reagan_speech_tkns_NoSigPunt)
```

### Finalmente cálculo de la frecuencia de las palabras, tras la individualización del texto en tokens y la eliminación de los signos de puntuación

```python
from nltk import FreqDist

freqDist_words = FreqDist(Reagan_speech_tkns_NoSigPunt)
```

```python
# Observación de las 15 palabras más frecuentes y el número de estas

freqDist_words_most=freqDist_words.most_common(15) 
freqDist_words_most
```

#### Visualización gráfica de las palabrás más frecuentes (15 primeras)

```python
# Conversión a una serie Pandas para graficar 
import pandas as pd
freqDist_words_most = pd.Series(dict(freqDist_words_most))

# Creación del gráfico
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10,5))

sns.barplot(x=freqDist_words_most.index, y=freqDist_words_most.values, ax=ax)
plt.title('Frecuencias de Palabras', size=13) 
plt.xlabel('Palabra')  
plt.ylabel('Frecuencia') 
plt.xticks(rotation=30)
plt.grid()
```

# Ejercicio 2
## Sacar los stopwords y realizar stemming al texto


```python
# Stopwords para eliminar las palabras vacías y sin aporte semántico al texto
from nltk.corpus import stopwords

Reagan_speech_sin_stop = [tkn for tkn in Reagan_speech_tkns_NoSigPunt if tkn not in stopwords.words("english")]
print(Reagan_speech_sin_stop)
```

```python
#Stemming para dejar las palabras en su forma raíz
from nltk.stem import PorterStemmer

p_s = PorterStemmer()
Reagan_speech_sin_stop_raiz= [p_s.stem(tkn) for tkn in Reagan_speech_sin_stop]
print(Reagan_speech_sin_stop_raiz)
```

```python
# Observación de las 15 palabras más frecuentes y el número de estas

Reagan_speech_clean = FreqDist(Reagan_speech_sin_stop_raiz)
Reagan_freqDist=Reagan_speech_clean.most_common(15) 
Reagan_freqDist
```

#### Visualización de las palabras más comunes tras aplicar "stopwords" y "stemming" al texto

```python
# Conversión a una serie Pandas para graficar 
import pandas as pd
Reagan_freqDist = pd.Series(dict(Reagan_freqDist))

# Creación del gráfico
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10,5))

sns.barplot(x=Reagan_freqDist.index, y=Reagan_freqDist.values, ax=ax)
plt.title('Frecuencias de Palabras', size=13) 
plt.xlabel('Palabra')  
plt.ylabel('Frecuencia') 
plt.xticks(rotation=30)
plt.grid()
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=900, height=700, background_color='gray')
wordcloud.generate_from_frequencies(Reagan_speech_clean)

plt.figure(figsize=(8, 7))
plt.title('Frecuencias de palabras', size=13)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
